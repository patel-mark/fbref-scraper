```python
import pandas as pd
import os  # Required for directory handling

def scrape_fbref_premier_league_stats():
    # Target URL
    url = "https://fbref.com/en/comps/9/Premier-League-Stats"
    
    # Table names mapping (table index to descriptive filename)
    table_mapping = {
        2: "Squad_Standard_Stats",
        3: "Squad_Standard_Stats(opponent stats)",
        4: "Squad_Goalkeeping_Stats",
        5: "Squad_Goalkeeping_Stats(opponent stats)",
        6: "Squad_Advanced_Goalkeeping_Stats",
        7: "Squad_Advanced_Goalkeeping_Stats(opponent stats)",
        8: "Squad_Shooting_Stats",
        9: "Squad_Shooting_Stats(opponent stats)",
        10: "Squad_Passing_Stats",
        11: "Squad_Passing_Stats(opponent stats)",
        12: "Squad_Pass_Types_Stats",
        13: "Squad_Pass_Types_Stats(opponent stats)",
        14: "Squad_Goal_Shot_Creation_Stats",
        15: "Squad_Goal_Shot_Creation_Stats(opponent stats)",
        16: "Squad_Defensive_Actions_Stats",
        17: "Squad_Defensive_Actions_Stats(opponent stats",
        18: "Squad_Possession_Stats",
        19: "Squad_Possession_Stats(opponent stats",
        20: "Squad_Playing_Time_Stats",
        21: "Squad_Playing_Time_Stats(opponent stats)",
        22: "Squad_Miscellaneous_Stats",
        23: "Squad_Miscellaneous_Stats(opponent stats"
    }
    
    # Create directory if it doesn't exist
    output_dir = "premier_league_data"
    os.makedirs(output_dir, exist_ok=True)
    
    try:
        # Read all tables on the page
        tables = pd.read_html(url)
        
        # Process tables 3 through 24 (indices 2 through 23 in the list)
        for i in range(2, 24):
            if i in table_mapping:
                df = tables[i]
                filename = f"{output_dir}/{table_mapping[i]}.csv"  # Save in the subfolder
                df.to_csv(filename, index=False)
                print(f"Saved table {i+1} as '{filename}'")
                
        print(f"\nAll tables saved successfully in '{output_dir}/'!")
        
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function
scrape_fbref_premier_league_stats()
```
